## Se filtran los datos no numÃ©ricos, se transforma la variable categÃ³rica magType para distinguir entre magnitudes: 
'mb':0,'mwr':1,'mww':2,'ml':3,'md':4,'mb_lg':5

Buscar quÃ© significan

```python
def coding(col, codeDict):
  colCoded = pd.Series(col, copy=True)
  for key, value in codeDict.items():
    colCoded.replace(key, value, inplace=True)
  return colCoded
quakedf["new_magType"] = coding(quakedf["magType"], {'mb':0,'mwr':1,'mww':2,'ml':3,'md':4,'mb_lg':5})
display(pd.value_counts(quakedf["new_magType"]))
```
